## WHAT IS A DOCKER?

Docker is a computer program that performs oeprating system level virtualizations

Docker is tool designed to create,deploy and run applications by using containers

In simple words, docker is a software containerization platform, meaning you can build your application, package them along with their depdendencies into a container and then these containers can be easily shipped to run on other machines.

### WHAT ARE CONTAINERS AND IMAGES

Containers are created with the help of Docker software and containers are freshly installed operating systems with other applications that can run on other machines.

Image are lightweight, standalone,executable package of software thtat includes everything needed to run an application 

example : To run selenium tests in Container we need applications like selenium webdriver , python, browsers like chrome/firefox, chromedriver/geckodriver, etc. So when a container is created it is a freshely created with only OS
and if we try to schedule test run it will fail as it does not have the dependencies installed. These dependencies can be specified in image.

Analogy: Just like we need to create a instance of a class to access its members and functions, similarly in Docker world image will have all details of softwares, system settings.
To use this image one creates a instance of the image using container. 

OBJECT IS A INSTANCE OF A CLASS

THUS CONTAINER IS A RUN TIME INSTANCE OF AN IMAGE 




### WHAT IS A DOCKER FILE?

A docker file contains instructions to build a docker image.
These instruction list dependencies needed. example selenium webdriver, python,chrome, chromedriver,etc

When docker file is executed docker software will scanned the docker file and create a docker image. Once we have docker image we will create a container using docker commands and once container is ready we deploy docker image into that container and run the tests 



### HOW EXACTLY ARE VMs (Virtual Machines) different from Dockers


image.png


**Virtualisation** is process of importing guest operation system on top of  host operating system (It is os of physical machine like laptop /desktop)

In containers all the resources required for the container are provided by the host OS when a particular app needs it where as in Vms the resources are allocated by Host OS during initilaizatin of the VMs itself.

example : Lets say a physical laptop has 10 GB RAM and if 3 VMS are created on it with 3GB RAM each, then each VMs internally will be managing its RAM of 3 GB, with no interaction to host system. Infact this leads to performance issues as there are 4 machines running on one laptop.
In case of containers even though we have 10 GB ram and 6 containers ready it is not allocated at the start. Lets say we have App B from below diagram  running , the host OS will check how many containers are running and it finds only container App B running and allocates  entire 10 GB ram to it. Thus Container with  App B is almost like host OS and performance issues are not faced. When tests on Container app b is done the ram is released back to the host OS.

**Containerization** is more efficient as there is no guest OS resources here and utilises host's OS , shares relevant libraries and resources as and when needed unlike VMs. Thus are light weight and faster than VMs.
Containers are therefore smaller than VMs and enable faster startup  with better performances, less isolation and greater compatibility possible due to sharing of hosts kernel.

7. Currently no container is running

to download latest version of the image 

c:\> docker pull imagename:latest

9. To find all the images downloaded to your machine type the following command

c:\\> docker images

image.png

10. Next task is to deploy this image on the container using following command

c:\\> docker run -d -p 4444:4444 -v /dev/shm:/dev/shm selenium/standalone.chrome:latest 

-d runs the container in background (detached)
-p 4444:4444 maps the local port 4444 to the port 4444 used by the selenium webdriver in the container.
:latest - is the tag/version of the image to use.

Why is -v /dev/shm:/dev/shm or --shm-size 2g necessary? This is a known workaround to avoid the browser crashing inside a docker container, here are the documented issues for Chrome and Firefox. The shm size of 2gb is arbitrary but known to work well, your specific use case might need a different value, it is recommended to tune this value according to your needs. Along the examples -v /dev/shm:/dev/shm will be used, but both are known to work.
